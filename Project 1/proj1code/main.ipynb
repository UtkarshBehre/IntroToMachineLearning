### Importing Libraries, extracting and shuffling the WDBC dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

#Extract the Wisconsin Diagnostic Breast Cancer (WDBC) dataset provided in the file wdbc.dataset
dataset = pd.read_csv("wdbc.dataset", header=None)

# Since Malignant is represented by 1 and Benign is represented by 0, and M and B are the only letters just in the column 1,
# we replace their values in the dataset accordingly
dataset = dataset.replace('M',1)
dataset = dataset.replace('B',0)

# shuffling the dataset using pandas sample method so we can just partition based on rows later 
# ref: https://stackoverflow.com/questions/29576430/shuffle-dataframe-rows
dataset = dataset.sample(frac=1).reset_index(drop=True)

# getting dataset as array
dataset = dataset.values

# we will not be using the column 0 which is id later on, even though it is still in the dataset
```

### All Functions definitions are defined in the below cell that are directly used in later cells

```python
# Defining all the methods' definitions that will be used in later cells

# this method partitions the given dataset into training, validation and testing parts into 80:10:10 ratio
# then it divides the labels from features for each dataset and return them
# train_perc tells what percentage of the data is to be used for training (value must be between 0-1)
# validation_perc tells what percentage of the remaining data is to be used for validation (value must be between 0-1)
# remaining data will be taken as test data
def partition_into_train_validation_test(dataset, train_perc, validation_perc):
    rows, cols = dataset.shape
    # first we get partitioning row number for training set
    train_part = math.floor(rows*train_perc)
    # then we get partitioning row number for validation set
    valid_part = train_part + math.floor((rows-train_part)*validation_perc)
    
    #then we divide the whole dataset using above partitioning row numbers we found
    train_dataset = dataset[:train_part]
    valid_dataset = dataset[train_part:valid_part]
    test_dataset = dataset[valid_part:]

    # next we set the features and labels for each training, validation, and test datasets to be returned
    # we are transposing the matrices using .T in order to make the multiplications and dot products work later on
    train_ftrs = train_dataset[:,2:cols].T
    train_lbls = train_dataset[:,1:2].T

    valid_ftrs = valid_dataset[:,2:cols].T
    valid_lbls = valid_dataset[:,1:2].T
    
    test_ftrs = test_dataset[:,2:cols].T
    test_lbls = test_dataset[:,1:2].T
    
    # return the training features, training labels, validation features, validation labels, 
    # testing features, and testing labels
    return train_ftrs, train_lbls, valid_ftrs, valid_lbls, test_ftrs, test_lbls

# This method normalizes the features of our dataset that is passed to it
def normalize(X):
    # X here has 30 rows (representing each feature) and columns(representing samples) depending on which 
    # set is passed(train, validation, or test)
    for row in range(X.shape[0]):
        # for each feature row we find min and max of the feature
        x_min = np.amin(X[row])
        x_max = np.amax(X[row])
        for col in range(X.shape[1]):
            # then we modify each feature value to make it lie between 0-1 
            # relative to the other values of the feature in the sample
            val = X[row][col]
            X[row][col] = (val-x_min)/(x_max-x_min)

# sigmoid function computes z and outputs values between 0 and 1 that is used to classify the class
# if output >=0.5 we call it as class 1 otherwise we say it should be of class 0
def sigmoid(z):
    return 1/(1 + np.exp(-z))

# function takes featues as X and respected labels along with
# epochs and learning_rate to be applied to the gradient descent
# draw is a boolean that decides whether to plot the metrics graphs or not
# X is of shape 30Xsample_size, where 30 is number of features so each row will have all values of that feature for all samples
# referred 1.1.2 Python+MLFrameworks.pdf from course documents of week1
def logistic_regression(X, Y, epochs, learning_rate, valid_ftrs, valid_lbls, draw=True):
    
    # We initialize different lists to record the changes in loss, accuracy, precision and recall 
    # for both training and validation dataset. We will later use these to plot graphs of each
    training_loss = []
    training_accuracy = []
    training_precision = []
    training_recall = []
    
    validation_loss = []
    validation_accuracy = []
    validation_precision = []
    validation_recall = []
    
    # m is total number of training sample cancer data available of patients
    # w is weight vector that decides importance of all features (initiallized using numpy)
    m = X.shape[1]
    w = np.zeros((X.shape[0],1))
    w0 = 0 # this is extra weight/ or the bias unrelated to any features
    
    # we go through entire training set(to improve) and validation set(to validate) every epoch 
    # but consider only training set while updating the weights
    for epoch in range(epochs):
        z = np.dot(w.T, X) + w0  # hypothesis function
        pY = sigmoid(z)  # calling sigmoid function on hypothesis function to get prediction
        
        # Formulas used together: 
        # Cost = -(-ylog pY + (1-y).log(1-y))
        # J(w) = (1/m).Summation(Cost(xi,yi))
        cost = -np.sum(np.multiply(np.log(pY), Y) + np.multiply((1-Y), np.log(1-pY)))/m
        
        # append cost value for training set every epoch for training loss graph
        training_loss.append(np.squeeze(cost))
        
        # append cost value for validation set every epoch for validation loss graph
        valid_z = np.dot(w.T,valid_ftrs) + w0
        valid_pY = sigmoid(valid_z)
        valid_cost = -np.sum(np.multiply(np.log(valid_pY), valid_lbls) + np.multiply((1-valid_lbls), np.log(1-valid_pY)))/m
        validation_loss.append(np.squeeze(valid_cost))
        
        #simultaneously update w and w0 using only training set evaluation, and the learning rate passed to this function
        dz = pY-Y
        dw = (1/m) * np.dot(X, dz.T)
        dw0 = (1/m) * np.sum(dz)
        w = w - learning_rate * dw # this updates the w vector
        w0 = w0 - learning_rate * dw0 # this updates the bias
        
        # we use predict() function to get predicted Y values being 0 or 1 (benign or Malignant)
        predicted_Y = classify(pY)
        # then we find true positives, true negatives, false positives and false negatives using evaluate_prediction() function
        TP, TN, FP, FN = evaluate_prediction(predicted_Y, Y, print_apr=False)
        # and we use that to get and append the current accuracy, precision, and recall
        training_accuracy.append(find_accuracy(TP,TN,FP,FN))
        training_precision.append(find_precision(TP,FP))
        training_recall.append(find_recall(TP, FN))
        
        # we do the same as we did for training set to get accuracy, precision, and recall for validation set
        predicted_Y = classify(valid_pY)
        TP, TN, FP, FN = evaluate_prediction(predicted_Y, valid_lbls, print_apr=False)
        validation_accuracy.append(find_accuracy(TP,TN,FP,FN))
        validation_precision.append(find_precision(TP,FP))
        validation_recall.append(find_recall(TP,FN))
    
    plt.plot(training_loss)
    plt.xlabel("Epochs")
    plt.ylabel("Training Loss")
    plt.title("Training Loss vs Epochs")
    # based on whether draw was passed as true or not we plot the graphs using draw_metrics_graph() function
    if draw == True:
        draw_metrics_graph(training_loss, training_accuracy, training_precision, training_recall, validation_loss, validation_accuracy, validation_precision, validation_recall)
    # we return all the final weights at the end of training
    return w, w0;

# this function takes lists representing loss, accuracy, precision, and recall for both training and validation 
# and plots the respective graphs in a 4x2 subplot figure
def draw_metrics_graph(train_loss, train_accuracy, train_precision, train_recall, valid_loss, valid_accuracy, valid_precision, valid_recall):
    
    # referred the matplotlib documentation, and modified few values as per the need
    # https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.subplots.html
    fig, graphs = plt.subplots(4,1, figsize=(8,20))
    fig.subplots_adjust(left=.2, bottom=.01, wspace=.5)
    fig.align_labels()
    
    # for each graph since we had to set labels, titles, and values repeatedly,
    # so set_graph_details() function is used to avoid code redundancy
    graph_tl = graphs[0]
    set_graph_details(graph_tl, train_loss, valid_loss, "Epochs", "Loss", "Loss vs Epochs", ("Training Loss", "Validation Loss"))
    graph_ta = graphs[1]
    set_graph_details(graph_ta, train_accuracy, valid_accuracy, "Epochs", "Accuracy (in %age)", "Accuracy vs Epochs", ("Training Accuracy", "Validation Accuracy"))
    graph_tp = graphs[2]
    set_graph_details(graph_tp, train_precision, valid_precision, "Epochs", "Precision (in %age)", "Precision vs Epochs", ("Training Precision", "Validation Precision"))
    graph_tr = graphs[3]
    set_graph_details(graph_tr, train_recall, valid_recall, "Epochs", "Recall (in %age)", "Recall vs Epochs", ("Training Recall" , "Validation Recall"))

# takes graph, and the details to set to the graph and sets them up
def set_graph_details(graph, values1, values2, xlbl, ylbl, title, legends):
    graph.set_xlabel(xlbl)
    graph.set_ylabel(ylbl)
    graph.set_title(title)
    graph.plot(values1)
    graph.plot(values2)
    graph.legend(legends)

# predicts the actual class being 0 or 1 taking set of features and corresponding weights
def predict(features, w, w0):
    z = np.dot(w.T, features) + w0
    y = sigmoid(z)
    return classify(y)

# classifies the prediction between 0-1 to a class 0 or 1
def classify(y):
    y[y>=0.5] = 1
    y[y<0.5] = 0
    return y

# Referred a site to understand calculation of true positives, true negatives, false positives, and false negatives
# ref: https://towardsdatascience.com/accuracy-paradox-897a69e2dd9b to learn about calculating TP, TN, FP, FN
def evaluate_prediction(predicted,actuals, print_apr):
    TP = 0
    TN = 0
    FP = 0
    FN = 0
    for i in range(predicted.shape[1]):
        if predicted[0][i] == actuals[0][i]:
            # actually has cancer and predicted having cancer
            if actuals[0][i] == 1:
                TP = TP + 1
            else: # actually doesn't have cancer and predicted to not have cancer
                TN = TN + 1
        else:
            # actually doesn't have cancer but predicted to have cancer
            if actuals[0][i] == 0:
                FP = FP + 1
            else: # actually have cancer but predicted to not have cancer
                FN = FN + 1
    # only prints the metrics if print_apr was passed true
    if print_apr:
        print_accuracy_precision_recall(TP,TN,FP,FN)
    return TP, TN, FP, FN

# simple function that find and prints accuracy, precision, and recall of a result
def print_accuracy_precision_recall(TP,TN,FP,FN):
    accuracy = find_accuracy(TP,TN,FP,FN)
    precision = find_precision(TP,FP)
    recall = find_recall(TP,FN)
    print("The performance of the algorithm is as belows:")
    print("Accuracy: %.2f%%" % accuracy)
    print("Precision: %.2f%%" % precision)
    print("Recall: %.2f%%" % recall)

# returns accuracy using accuracy formula
def find_accuracy(TP, TN, FP, FN):
    return (TP+TN)*100/(TP+TN+FP+FN)

# returns precision using precision formula
def find_precision(TP,FP):
    return TP*100/(TP+FP)

# returns recall using recall formula
def find_recall(TP,FN):
    return TP*100/(TP+FN)

# plots the graph that shows the importance of the features used in predicting the final class
def plot_feature_importance_graph(w):
    ftrs_cols = []
    for i in range(30):
        ftrs_cols.append(i+2)

    # ref: https://pythonspot.com/matplotlib-bar-chart/
    y_pos = np.arange(len(ftrs_cols))
    plt.bar(y_pos, w[:,0], align='center', alpha=0.5, width = 0.4)
    plt.xticks(y_pos, ftrs_cols)
    plt.ylabel('Weight (Importance)')
    plt.xlabel('Feature Column Number(Based on original dataset)')
    plt.title('Importance of features')
    # resize the figure to make it more readable
    # ref: https://codeyarns.com/2014/10/27/how-to-change-size-of-matplotlib-plot/
    fig_size = plt.rcParams["figure.figsize"]
    plt.rcParams["figure.figsize"] = [10,6]
```

### Normalization and  partitioning of data

```python
# partitioning the dataset into training, validation, and testing 
# 80% of the dataset is used for training, 10% for validation, and remaining 10% for testing
# the way partitioning is done is, first 80% and 20% are seperated, 
# and then 50% of the 20% is taken for validation and remaining 50% for testing

# normalizing all the features of the dataset
normalize(dataset[:,2:].T)

train_ftrs, train_lbls, valid_ftrs, valid_lbls, test_ftrs, test_lbls = partition_into_train_validation_test(dataset, .8, .5)

# printing the shapes of features and labels for better understanding of the dimensions of the matrices
# before further doing any further computation
shapes = [["Training Features", train_ftrs.shape],["Training Labels", train_lbls.shape],["Validation Features", valid_ftrs.shape],["Validation Labels", valid_lbls.shape],["Testing Features", test_ftrs.shape],["Testing Labels", test_lbls.shape]]
print(shapes)
```

### Training the model using epochs and learning rate hyperparameters

```python
# Here we can do the training and evaluation by adjusting the epochs, and learning rate hyper parameters
# and observing the training and validation loss, accuracy, precision, and recall progression every epoch

w, w0 = logistic_regression(train_ftrs, train_lbls, 3000, .09,  valid_ftrs, valid_lbls, True)
```

### Testing our model on the testing dataset and printing the final results

```python
# Here we do the final testing on the remaining test dataset using the weights and bias we achieved by training
# and we get the final accuracy, precision and recall of the results predicted by the algorithm

predicted_lbls = predict(test_ftrs, w, w0)
evaluate_prediction(predicted_lbls, test_lbls, print_apr=True)
```

### Plotting a bar graph for importance of the features based on their final weight values

```python
# Finally we plot the importance of the given features based on the respective weights we found after training
plot_feature_importance_graph(w)
```
